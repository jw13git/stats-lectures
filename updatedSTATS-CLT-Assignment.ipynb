Part 1: Explore the Beta Distribution
The beta distribution has two parameter, a and b.

Set a=1, b=1 and use stats.beta.rvs to generate size=20, size=50, size=1000, and size=100,000 random variables from the resulting beta distribution. 

Graph using seaborn's distplot function. What do you think the min, max, mean, and median of the beta distribution are (with a=1, b=1) as size gets infinitely large?

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import beta

from collections import Counter
```

```python
a,b = 1,1
r = beta.rvs(a, b, size=20)
r1 = beta.rvs(a,b,size=50)
r2 = beta.rvs(a,b,size=1000)
r3 = beta.rvs(a,b,size=100000)
```

```python
sns.distplot(r)
```

```python
sns.distplot(r1)
```

```python
sns.distplot(r2)
```

```python
sns.distplot(r3)
```

```python
# What do you think the min, max, mean, and median of the beta distribution are (with a=1, b=1) as size gets infinitely large?
print(f"\n min={np.min(r)}\n max={np.max(r)}\n median={np.median(r)}\n mean={np.mean(r)}\n std={np.std(r, ddof=1)}\n" )

print(f"\n min={np.min(r1)}\n max={np.max(r1)}\n median={np.median(r1)}\n mean={np.mean(r1)}\n std={np.std(r1, ddof=1)}\n" )

print(f"\n min={np.min(r2)}\n max={np.max(r2)}\n median={np.median(r2)}\n mean={np.mean(r2)}\n std={np.std(r2, ddof=1)}\n" )

print(f"\n min={np.min(r3)}\n max={np.max(r3)}\n median={np.median(r3)}\n mean={np.mean(r3)}\n std={np.std(r3, ddof=1)}\n" )
```

```python
a,b = 20,20
r4 = beta.rvs(a, b, size=20)
r5 = beta.rvs(a,b,size=50)
r6 = beta.rvs(a,b,size=1000)
r7 = beta.rvs(a,b,size=100000)
```
